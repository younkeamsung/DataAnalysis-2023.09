## Matplotlib 한글 사용 - VSCode

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

- 한글 설정 이전

```python
plt.plot([0,1,4,9], label='x의 제곱')     # X는 생략, Y 값이 4개니까 x[0,1,2,3], y[0,1,4,9]
plt.title('라인 플롯')
plt.xlabel('X축'), plt.ylabel('Y축')
plt.grid()
plt.legend()
plt.show()
```

- 윈도우에 설치되어 있는 폰트 이용
