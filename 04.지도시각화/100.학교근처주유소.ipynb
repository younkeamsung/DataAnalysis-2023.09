```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('data/학교근처주유소2.csv')
df
```

```python
map = folium.Map(location=[df.위도.mean(),df.경도.mean()], zoom_start=14)
for i in df.index:
    folium.Marker(
        location=[df.위도[i], df.경도[i]],    
        tooltip=df.이름[i],
        popup=folium.Popup(df.주소[i], max_width=200)
    ).add_to(map)

folium.Marker(
    location=[df.위도[0],df.경도[0]],
    icon=folium.CustomIcon('data/GS칼텍스.jpg', icon_size=(30,30))
).add_to(map)

title_html = '<h3 align="center" style="font-size:20px">학교근처주유소</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
