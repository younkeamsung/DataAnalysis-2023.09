```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.tail(3)
```

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
```

```python
bins = [0, 19, 30, 50, 70, 100]
titanic['age_group'] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail() 
```

```python
titanic['category3'] = \
    titanic.apply(lambda x: x.age_group + ' 남성' if (x.age >= 20) & (x.sex == 'male') else \
                            x.age_group + ' 여성' if (x.age >= 20) & (x.sex == 'female') else \
                                '미성년자' if x.age < 20 else np.nan, axis=1)
titanic.head(10)
```

```python
# age의 결측치를 평균으로 대체
titanic.age.fillna(titanic.age.mean().round(2), inplace=True)
titanic['age_group'] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail()
```

```python
# male/female --> 남성/여성
titanic['성별'] = titanic.sex.apply(lambda x: ' 남성' if x == 'male' else ' 여성')
titanic.tail()
```

```python
titanic['category3'] = titanic.apply(
    lambda x: x.age_group + x.성별 if x.age >= 20 else '미성년자', axis=1)
titanic.tail(10)
```
